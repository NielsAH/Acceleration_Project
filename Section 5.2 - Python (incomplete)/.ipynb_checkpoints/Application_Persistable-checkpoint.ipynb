# Persistable for thesis

We start with some general imports.

```python
import persistable

import numpy as np
import matplotlib.pyplot as plt
import matplotlib

import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.datasets import make_blobs

import plotly.express as px

from urllib.request import urlopen
from io import BytesIO
```

## Trying out to see if Persistable works

```python
np.random.seed(12) 
cluster1 = np.random.normal(loc=[0.4,0], scale=0.1, size=(800, 2))
cluster2 = np.random.normal(loc=[0,0], scale=0.1, size=(800, 2))
data = np.concatenate((cluster1, cluster2), axis=0)

plt.figure(figsize=(10,10))
plt.scatter(data[:,0], data[:,1], alpha=0.5)
plt.show()
```

```python
p = persistable.Persistable(data, n_neighbors="all")
pi = persistable.PersistableInteractive(p)
pi.start_ui()
```

```python
#localhost:8059
```

```python
cluster_labels = pi.cluster()
```

```python
cluster_labels
```

```python
cNorm  = matplotlib.colors.Normalize(vmin=0, vmax=np.amax(cluster_labels))
scalarMap = matplotlib.cm.ScalarMappable(norm=cNorm, cmap='Paired')
noise_color = (211/255, 211/255, 211/255, 1) # light grey

# plot data with clusters indicated by colors
plt.figure(figsize=(10,10))
plt.scatter(data[:,0], data[:,1], c=[noise_color if x==-1 else scalarMap.to_rgba(x) for x in cluster_labels], alpha=0.5)

plt.show()
```

## Applying Persistable to an imbedding of a step distance matrix into Euclidean space

Note: Persistable uses coordinates, not a dissimilarity matrix. For our purposes in the thesis we have used the visualisations using Persistable, using coordinates in the highest Euclidean space that contributed anything (often around dim 200 for 500 steps). However, all clusters made using the Persistable idea in R etc. use the origianl data, not the dissimilarity matrix. A variant of Persistable in R using the dissimilarity matrix can be found in the Misc. folder and gave the same results, but was not fully completed.

```python
df = pd.read_csv('name.csv')
```

```python
data = df.to_numpy()
```

```python
p = persistable.Persistable(data, n_neighbors="all")
pi = persistable.PersistableInteractive(p)
pi.start_ui()
```

```python
cluster_labels = pi.cluster()
```

```python
pca = PCA(n_components=3)
pca_result = pca.fit_transform(df)

# Create a DataFrame for plotting
plot_df = pd.DataFrame(pca_result, columns=['PC1', 'PC2', 'PC3'])
plot_df['cluster'] = cluster_labels

# Plot with plotly
fig = px.scatter_3d(plot_df, x='PC1', y='PC2', z='PC3', color=plot_df['cluster'].astype(str))
fig.update_traces(marker=dict(size=3))  # Adjust size as needed

fig.show()
```

## Concentric dataset and variable noise

```python
def concentric_data_2D_eqpoints(N, radii, noise):
    data = np.empty((0,2))
    for r in radii:
        t = np.linspace(0, 2*3.1415,N , endpoint=False)
        temp = np.transpose([(r)*np.cos(t), (r)*np.sin(t)]) + noise*np.random.random((N,2))
        data = np.concatenate((data,temp))
    plt.show()
    return data
```

```python
def concentric_data_2D_eqdist(N, radii, noise):
    data = np.empty((0,2))
    for r in radii:
        t = np.linspace(0, 2*3.1415,int(np.floor(N*r)), endpoint=False)
        temp = np.transpose([(r)*np.cos(t), (r)*np.sin(t)]) + noise*np.random.random((int(np.floor(N*r)),2))
        data = np.concatenate((data,temp))
    return data
```

```python
np.random.seed(12) 
data = concentric_data_2D_eqdist(300, [1,2,3,4,5], 0.5)
plt.scatter(data[:, 0], data[:, 1])
plt.axis('equal')
plt.show()
```

```python
np.random.seed(12) 
data = concentric_data_2D_eqdist(1500, [1,1/2,1/3,1/4,1/5], 0.019)
plt.scatter(data[:, 0], data[:, 1])
plt.axis('equal')
plt.show()
```

```python
np.random.seed(12) 
data = concentric_data_2D_eqdist(300, [1,2,3,4,5], 0.65)
plt.scatter(data[:, 0], data[:, 1])
plt.axis('equal')
plt.show()
```

```python
p = persistable.Persistable(data, n_neighbors="all")
pi = persistable.PersistableInteractive(p)
port = pi.start_ui()
```

```python
cluster_labels = pi.cluster()
```

```python
cNorm  = matplotlib.colors.Normalize(vmin=0, vmax=np.amax(cluster_labels))
scalarMap = matplotlib.cm.ScalarMappable(norm=cNorm, cmap='Paired')
noise_color = (211/255, 211/255, 211/255, 1) # light grey

# plot data with clusters indicated by colors
plt.figure(figsize=(10,10))
plt.scatter(data[:,0], data[:,1], c=[noise_color if x==-1 else scalarMap.to_rgba(x) for x in cluster_labels], alpha=0.5)

plt.show()
```

## Hierarchical data

In this application, we work on a hierachical type of dataset. This means that different density scales give different seperations in clusters. We use this to see how easy it is for the user to find different parameters to get the different levels. 

```python
from sklearn.datasets import make_blobs, make_classification

X1, Y1 = make_blobs(n_samples=1000,n_features=2,
centers=10, cluster_std=1.2,
center_box=(-10.0, 10.0),
shuffle=True, random_state=12)
plt.scatter(X1[:, 0], X1[:, 1], marker="o", s=25, edgecolor="k")
plt.show()
```

```python
app3 = persistable.Persistable(X1, n_neighbors="all")
piapp3 = persistable.PersistableInteractive(app3)
portapp3 = piapp3.start_ui()
```

```python
cluster_labels = piapp3.cluster()

# choose color map
import matplotlib
cNorm  = matplotlib.colors.Normalize(vmin=0, vmax=np.amax(cluster_labels))
scalarMap = matplotlib.cm.ScalarMappable(norm=cNorm, cmap='Paired')
noise_color = (211/255, 211/255, 211/255, 1) # light grey

# plot data with clusters indicated by colors

plt.scatter(X1[:,0], X1[:,1], c=[noise_color if x==-1 else scalarMap.to_rgba(x) for x in cluster_labels], alpha=0.5)

plt.show()
```
